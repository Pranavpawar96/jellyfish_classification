```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
file_path = 'C:/Users/Admin/Downloads/Quote-Equity-TATAMOTORS-EQ-13-09-2023-to-13-09-2024.csv'
data = pd.read_csv(file_path)
```

```python
# Check column names to confirm the correct name for the 'Date' column
data.columns
```

```python
# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Re-run the code to remove commas and convert columns to numeric types
cols_to_clean = ['OPEN', 'HIGH', 'LOW', 'PREV. CLOSE', 'ltp', 'close', 'vwap', 'VOLUME', 'VALUE', 'No of trades']
for col in cols_to_clean:
    data[col] = data[col].astype(str).str.replace(',', '').astype(float)
```

```python
# Sort by date in ascending order for plotting
data = data.sort_values(by='Date')

# Plot Closing Price Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['close'], label='Closing Price', color='blue')
plt.title('Daily Closing Price of TATAMOTORS')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot Volume Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['VOLUME'], label='Volume', color='orange')
plt.title('Daily Trading Volume of TATAMOTORS')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid()
plt.show()
```
